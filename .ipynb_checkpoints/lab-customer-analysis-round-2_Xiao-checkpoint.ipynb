```python
#Check and deal with NaN values
df.isna().sum()
```

```python
df['state'].value_counts()
```

```python
df['state']=df['state'].fillna('California')
```

```python
df['response'].value_counts()
```

```python
df['response']=df['response'].fillna('No')
```

```python
df['vehicle class'].value_counts()
```

```python
df['vehicle class']=df['vehicle class'].fillna('Four-Door Car')
```

```python
df['vehicle size'].value_counts()
```

```python
df['vehicle size']=df['vehicle size'].fillna('Medsize')
```

```python
df['vehicle type'].value_counts()
```

```python
df['vehicle type']=df['vehicle type'].fillna('A')
```

```python
df.isna().sum()
```

```python
df['months since last claim'].mean()
```

```python
df['months since last claim'] = df['months since last claim'].fillna(15.149070740488469)
```

```python
df['number of open complaints'].mean()
```

```python
df['number of open complaints'] = df['number of open complaints'].fillna(df['number of open complaints'].mean())
```

```python
df.isna().sum()
```
